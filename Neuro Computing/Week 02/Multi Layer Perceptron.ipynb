# 1. Lakukan Multi Perceptron untuk data XOR seperti assignment di atas

```python
dataset_xor = pd.read_csv('../Dataset/XOR.csv')
data_xor = dataset_xor.iloc[:, 0:-1]
label_xor = dataset_xor.iloc[:, -1]

iter = np.array([10, 20, 30, 40, 50, 100, 150 , 200, 250, 300])

def mse(actual, pred):
    actual, pred = np.array(actual), np.array(pred)
    return np.square(np.subtract(actual, pred)).mean()

error = []

for i in range(len(iter)):
    model = MLP(hidden_layer_sizes=(7), max_iter=iter[i],
                activation='logistic', learning_rate_init=0.1)
    model.fit(data_xor, label_xor)
    pred_xor = model.predict(data_xor)
    error.append(mse(label_xor, pred_xor))
    print('Score:', model.score(data_xor, label_xor))

plt.plot(iter, error)
plt.xlabel('Iteration No.')
plt.ylabel('Mean Squared Error')
plt.title('Gradient Descent XOR')
plt.show()
```
