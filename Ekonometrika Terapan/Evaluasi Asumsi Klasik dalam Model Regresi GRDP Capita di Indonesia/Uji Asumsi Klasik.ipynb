# Preprocessing
## Standardization

```python
from sklearn.preprocessing import StandardScaler

dt = dataset.copy()

X = dt.drop(['Province', 'Year'], axis = 1)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled_df = pd.DataFrame(X_scaled, columns = X.columns)

data_scaled_df = pd.concat([dataset[['Province', 'Year']], X_scaled_df], axis=1)
data_scaled_df = data_scaled_df[['Province', 'Year'] + list(X.columns)]

data_scaled_df
```

## Correlation Matrix

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (8, 6))

corr_matrix = data_scaled_df.iloc[:, 2:].corr()
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm', annot_kws = {"size": 7},
            fmt = ".2f", linewidths = 0.5, cbar_kws={'shrink': 0.7})

plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()

plt.show()
```

### Pemisahan Variabel Target (y)

```python
import statsmodels.api as smf

X = data_scaled_df.iloc[:, 3:]
y = data_scaled_df['GRDP Capita']

# OLS Model
model = smf.OLS(y, X).fit()
model.summary()
```

```python
y_pred = model.predict(X) 
error = y - y_pred 
gap = pd.DataFrame({'Actual': y,
                    'Predicted': y_pred,
                    'Error':error}) 
gap.head(10)
```

```python
from sklearn import metrics
import numpy as np

mae = metrics.mean_absolute_error(y, y_pred) 
mse = metrics.mean_squared_error(y, y_pred) 
rmse = np.sqrt(metrics.mean_squared_error(y, y_pred)) 
print(f'Mean Absolute Error: {mae:.5f}') 
print(f'Mean Square Error: {mse:.5f}') 
print(f'Root Mean Square Error: {rmse:.5f}')
```

# Uji Asumsi Klasik
## A. Multikolinearitas

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data = pd.DataFrame() 
vif_data["feature"] = X.columns 
vif_data["VIF"] = [variance_inflation_factor(X.values, i) 
    for i in range(len(X.columns))] 

print(vif_data)
```

### Solusi 1: PCA

```python
from sklearn.decomposition import PCA 
from sklearn.preprocessing import scale,StandardScaler

pca = PCA() 
xr=pca.fit_transform(scale(X))
```

```python
xr = pd.DataFrame(xr) 
vif_data_r = pd.DataFrame() 
vif_data_r["feature"] = X.columns 
vif_data_r["VIF"] = [variance_inflation_factor(xr.values, i) 
    for i in range(len(xr.columns))] 

print(vif_data_r)
```

```python
xr = smf.add_constant(xr)
modelPCA= smf.OLS(y, xr).fit()
modelPCA.summary()
```

### Solusi 2: Ridge Regression

```python
alphas = 10 ** np.linspace(10,  -2,100) * 0.5 
alphas
```

```python
from sklearn.linear_model import Ridge, RidgeCV, Lasso, LassoCV 

scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
ridge = Ridge()
coefs = []

for a in alphas:
    ridge.set_params(alpha=a)
    ridge.fit(x_scaled, y)  
    coefs.append(ridge.coef_)
print(np.shape(coefs))
```

```python
ax = plt.gca() 
ax.plot(alphas, coefs) 
ax.set_xscale('log') 
plt.axis('tight') 
plt.xlabel('alpha') 
plt.ylabel('weights')
```

```python
alphas = [None]

for alpha in alphas:
    if alpha is None:
        ridge = RidgeCV(alphas = [0.1, 0.5, 1, 2, 4, 10])  
        ridge.fit(x_scaled, y)
        best_alpha = ridge.alpha_ 
        print(f"Best Alpha (RidgeCV): {best_alpha}")
    else:
        ridge = Ridge(alpha=alpha)
        ridge.fit(x_scaled, y)
        best_alpha = alpha
    pred_ridge = ridge.predict(x_scaled)
    print(f"Mean Squared Error: {metrics.mean_squared_error(y, pred_ridge)}") 
```

```python
pd.Series(ridge.coef_, index=X.columns)
```

### Solusi 3: LASSO Regression

```python
alphas = 10 ** np.linspace(10, -2, 100) * 0.5 
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)

lasso = Lasso(max_iter=10000)
coefs = [] 
for a in alphas: 
    lasso.set_params(alpha=a) 
    lasso.fit(scale(X), y) 
    coefs.append(lasso.coef_) 
ax = plt.gca() 
ax.plot(alphas*2, coefs) 
ax.set_xscale('log')
plt.axis('tight') 
plt.xlabel('alpha') 
plt.ylabel('weights')
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)

lasso_cv = LassoCV(alphas=None, cv=10, max_iter=100000)
lasso_cv.fit(x_scaled, y)
best_alpha = lasso_cv.alpha_
lasso = Lasso(alpha=best_alpha, max_iter=10000)
lasso.fit(x_scaled, y)
mse = metrics.mean_squared_error(y, lasso.predict(x_scaled))

print(f"Best Alpha (Lasso): {best_alpha}")
print(f"Mean Squared Error: {mse}")
```

## B. Uji Heteroskedastisitas

```python
from statsmodels.stats.diagnostic import het_breuschpagan, het_white
from statsmodels.compat import lzip 

bp_test = het_breuschpagan(modelPCA.resid, modelPCA.model.exog)
bp_labels = ['Statistik LM', 'P-Value LM', 'Statistik F', 'P-Value F']
lzip(bp_labels, bp_test)
```

```python
white_test = het_white(modelPCA.resid, modelPCA.model.exog)
white_labels = ['Statistik LM', 'P-Value LM', 'Statistik F', 'P-Value F']
lzip(white_labels, white_test)
```

```python
from sklearn.model_selection import train_test_split
from statsmodels.stats.stattools import durbin_watson
from scipy.stats import boxcox

X = data_scaled_df.iloc[:,3:]
y = data_scaled_df['GRDP Capita']

Q1 = X.quantile(0.25)
Q3 = X.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

X_cleaned = X[~((X < lower_bound) | (X > upper_bound)).any(axis=1)]
y_cleaned = y[X_cleaned.index]

if (y_cleaned <= 0).any():
    y_cleaned += (abs(y_cleaned.min()) + 1)

y_boxcox, lambda_boxcox = boxcox(y_cleaned)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cleaned)
```

```python
import statsmodels.api as sm

X_train_scaled, X_test_scaled, y_train_boxcox, y_test_boxcox = train_test_split(X_scaled, y_boxcox, test_size=0.2, random_state=42)

X_train_with_const = sm.add_constant(X_train_scaled)

modelWLS = sm.WLS(y_train_boxcox, X_train_with_const).fit()

residuals = modelWLS.resid

_, pval_bp, _, _ = het_breuschpagan(residuals, modelWLS.model.exog)

print('Breusch-Pagan Test Statistics WLS:')
print(f"P-Value Breusch-Pagan: {pval_bp}")

_, pval_white, _, _ = het_white(residuals, modelWLS.model.exog)

print("\nWhite Test Statistics WLS:")
print(f"P-Value White Test: {pval_white}")
```
